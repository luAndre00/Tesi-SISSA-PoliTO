```python
def carica_variabile(file):
    with open(file, 'r') as f:
        return eval(f.read())

# Chiama la funzione per caricare la variabile
variabile_caricata = carica_variabile('loss_learningRate.txt')
print(type(variabile_caricata))

import matplotlib.pyplot as plt 
plt.loglog(variabile_caricata)
plt.gcf().savefig('LR_loss.pdf', format='pdf')
```
