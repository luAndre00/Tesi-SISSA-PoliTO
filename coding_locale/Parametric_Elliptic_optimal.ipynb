```python
import argparse
import numpy as np
import torch
import matplotlib.pyplot as plt
from torch.nn import Softplus
from torch.optim.lr_scheduler import ReduceLROnPlateau

from pina import LabelTensor
from pina.solvers import PINN
from pina.model import MultiFeedForward
from pina.plotter import Plotter
from pina.trainer import Trainer
from pina.model import FeedForward
from pina.callbacks import MetricTracker

from pina import Condition
from pina.geometry import CartesianDomain
from pina.equation import SystemEquation, FixedValue
from pina.problem import SpatialProblem, ParametricProblem
from pina.operators import laplacian
```

```python
# ===================================================== #
#                                                       #
#  This script implements the two dimensional           #
#  Parametric Elliptic Optimal Control problem.         #
#  The ParametricEllipticOptimalControl class is        #
#  inherited from TimeDependentProblem, SpatialProblem  #
#  and we denote:                                       #
#           u --> field variable                        #
#           p --> field variable                        #
#           y --> field variable                        #
#           x1, x2 --> spatial variables                #
#           mu, alpha --> problem parameters            #
#                                                       #
#  More info in https://arxiv.org/pdf/2110.13530.pdf    #
#  Section 4.2 of the article                           #
# ===================================================== #


class ParametricEllipticOptimalControl(SpatialProblem, ParametricProblem):

    # setting spatial variables ranges
    xmin, xmax, ymin, ymax = -1, 1, -1, 1
    x_range = [xmin, xmax]
    y_range = [ymin, ymax]
    # setting parameters range
    a = 1
    # amin, amax = 0.01, 1
    mumin, mumax = 0.5, 3
    mu_range = [mumin, mumax]
    # a_range = [amin, amax]
    # setting field variables
    output_variables = ['u', 'p', 'y']
    # setting spatial and parameter domain
    spatial_domain = CartesianDomain({'x1': x_range, 'x2': y_range})
    parameter_domain = CartesianDomain({'mu': mu_range})
    # parameter_domain = CartesianDomain({'mu': mu_range, 'alpha': a_range})

    # equation terms as in https://arxiv.org/pdf/2110.13530.pdf
    #Questa è la prima equazione della equazione 4.9
    def term1(input_, output_):
        laplace_p = laplacian(output_, input_, components=['p'], d=['x1', 'x2']) #calcolo la derivata di p rispetto alle variabili x1 e x2 solamente
        return output_.extract(['y']) - input_.extract(['mu']) - laplace_p

    #Questa è la quarta equazione della equazione 4.9
    def term2(input_, output_):
        laplace_y = laplacian(output_, input_, components=['y'], d=['x1', 'x2'])
        return - laplace_y - output_.extract(['u'])

    #Questa è la quinta equazione della equazione 4.9
    def fixed_y(input_, output_):
        return output_.extract(['y'])

    # #Questa è la seconda equazione della equazione 4.9
    def fixed_p(input_, output_):
        return output_.extract(['p'])

    #Le condizioni al bordo sono abbastanza semplici in questo caso, poiché la variabile di stato y
    #è posta a zero su tutto il bordo, mentre la variabile lagrangiana aggiunta p (z nel paper), è presa in
    #H^1_0, quindi sempre nulla sul bordo. Difatti le ultime due def servono solo da mettere qua sotto.
    #A tal fine il metodo SystemEquation serve solo a dire che le due condizioni dentro devono valere
    #contemporaneamente, non a caso infatti si chiama SystemEquation

    # setting problem condition formulation
    conditions = {
        'gamma1': Condition(
            location=CartesianDomain({'x1': x_range, 'x2':  1, 'mu': mu_range}),
            equation=SystemEquation([fixed_y, fixed_p])),
        'gamma2': Condition(
            location=CartesianDomain({'x1': x_range, 'x2': -1, 'mu': mu_range}),
            equation=SystemEquation([fixed_y, fixed_p])),
        'gamma3': Condition(
            location=CartesianDomain({'x1':  1, 'x2': y_range, 'mu': mu_range}),
            equation=SystemEquation([fixed_y, fixed_p])),
        'gamma4': Condition(
            location=CartesianDomain({'x1': -1, 'x2': y_range, 'mu': mu_range}),
            equation=SystemEquation([fixed_y, fixed_p])),
        'D': Condition(
            location=CartesianDomain(
                {'x1': x_range, 'x2': y_range,
                'mu': mu_range
                }),
            equation=SystemEquation([term1, term2])), #Ho aggiunto qui term3
    }
```

Extra feature:

$$
k(x_0,x_1) = (1-x_0^2)(1-x_1^2)
$$

```python
torch.manual_seed(316680)

#In questo problema l'extra feature viene usata, è un po' strana
class myFeature(torch.nn.Module):
    """
    Feature: sin(x)
    """

    def __init__(self):
        super(myFeature, self).__init__()

    def forward(self, x):
        t = (-x.extract(['x1'])**2+1) * (-x.extract(['x2'])**2+1)
        # print(t)
        return LabelTensor(t, ['k0'])


class CustomMultiDFF(MultiFeedForward):

    def __init__(self, dff_dict):
        super().__init__(dff_dict)

    #Questo si usa al posto della equazione che "manca", altrimenti viene male
    def forward(self, x):
        # out = self.uu(x)
        # out.labels = ['u', 'y']
        # p = LabelTensor(
        #     (out.extract(['u']) * 1), ['p'])

        
        out = self.uu(x)
        out.labels = ['u', 'y']
        print(out)
        u = out.extract(['u'])*(1-x.extract(['x1'])**2)*(1-x.extract(['x2'])**2)
        y = out.extract(['y'])*(1-x.extract(['x1'])**2)*(1-x.extract(['x2'])**2)
        t = torch.cat((u,y), dim=1)

        strong_out = LabelTensor(t, ['u','y'])
        p = LabelTensor((strong_out.extract(['u'])), ['p'])
        print(p)
        return strong_out.append(p)


if __name__ == "__main__":

    parser = argparse.ArgumentParser(description="Run PINA")
    parser.add_argument("--load", help="directory to save or load file", type=str)
    parser.add_argument("--features", help="extra features", type=int, default = 1)
    parser.add_argument("--epochs", help="extra features", type=int, default=20)
    parser.add_argument('-f') #Serve per risolvere l'errore di sotto
    args = parser.parse_args()

    if args.features is None:
        args.features = 0

    # extra features
    feat = [myFeature()] if args.features else []
    args = parser.parse_args()

    # create problem and discretise domain, qua in sostanza genero i punti su cui voglio allenare
    opc = ParametricEllipticOptimalControl() #Creo un istanza di modello
    ne = 3
    opc.discretise_domain(n= ne, mode='random', variables=['x1', 'x2'], locations=['D'])
    opc.discretise_domain(n= ne, mode='random', variables=['mu'], locations=['D'])
    opc.discretise_domain(n= ne, mode='random', variables=['x1', 'x2'], locations=['gamma1', 'gamma2', 'gamma3', 'gamma4'])
    opc.discretise_domain(n= ne, mode='random', variables=['mu'], locations=['gamma1', 'gamma2', 'gamma3', 'gamma4'])

    # create model, qui è un po' più complesso e per costruire il modello si appoggia alla classe
    #che ha creato prima CustomMultiDFF
    model = CustomMultiDFF(
        {
            'uu': {
                'input_dimensions': 3 + len(feat), #due più due parametri
                'output_dimensions': 2,
                'layers': [40, 40, 20],
                'func': Softplus,
            },
        }
    )
    
    # create PINN
    pinn = PINN(problem=opc, model=model, optimizer_kwargs={'lr' : 0.002}, extra_features=feat)
    #aggiungere lo scheduler
    
    # print(pinn.extra_features)

    # create trainer
    directory = 'pina.parametric_optimal_control_{}'.format(bool(args.features))
    trainer = Trainer(solver=pinn, accelerator='cpu', max_epochs=args.epochs, default_root_dir=directory)

    if args.load:
        pinn = PINN.load_from_checkpoint(checkpoint_path=args.load, problem=opc, model=model, extra_features=feat)
        plotter = Plotter()
        plotter.plot(pinn, fixed_variables={'mu' : 1 , 'alpha' : 0.001}, components='y')
    else:
        trainer.train();
```
